## 데이터 셋 확인
**<변수 설명>**
1. ***Survived: 생존 여부 (integer)***
    - **Target 변수**로, 1 과 0 으로 표현된다.
2. ***Pclass: 티켓의 클래스 (integer)***
    - 1 = 1st, 2 = 2nd, 3 = 3rd 클래스로 나뉘는 범주형 변수이다.
3. ***Sex: 성별 (string)***
    - male 과 female 로 구분되는 이진 변수이다.
4. ***Age: 나이 (integer)***
    - Continuous 변수이다.
5. ***SibSp: 함께 탑승한 형제와 배우자의 수 (integer)***
    - Quantitative 변수이다.
6. ***Parch: 함께 탑승한 부모, 아이의 수 (integer)***
    - Quantitative 변수이다.
7. ***Ticket: 티켓 번호 (string)***
    - **알파벳 + integer** 로 구성되어 있다.
8. ***Fare: 탑승료 (float)***
    - Continuous 변수이다.
9. ***Cabin: 객실 번호 (string)***
    - **알파벳 + interger** 로 구성되어 있다.
10. ***Embarked: 탑승 항구 (string)***
    - C = Cherbourg, Q = Queenstown, S = Southampton 이렇게 3개로 구성되어 있다.

## Age에 따른 생존률의 차이 확인

```python
plt.figure(figsize = (10, 5))
age_hist = sns.distplot(df_train['Age'], label = 'Skewness: {:.2f}'.format(df_train['Age'].skew()))
age_hist = age_hist.legend(loc = 'best')
```

- **연속형 변수인 Age의 분포**를 확인해보니, **약간의 치우침**이 보인다.
- 나중에 Feature Engineering 과정에서 **Null 값이 존재해서 치우친 분포가 나온 것인지 확인**하고, **로그 변환** 수행 여부를 결정할 것이다.

```python
plt.figure(figsize = (9, 5))

sns.kdeplot(df_train[df_train['Survived'] == 1]['Age'], shade = True)
sns.kdeplot(df_train[df_train['Survived'] == 0]['Age'], shade = True)

plt.legend(['Suvived = 1', 'Survived = 0'])
plt.show()
```

- 젊은 연령대(20 ~ 30대)의 생존자들이 많은 것을 알 수 있다.

```python
cummulate_survival_ratio = []

for i in range(1, 81):
    cummulate_survival_ratio.append(df_train[df_train['Age'] < i]['Survived'].sum() / len(df_train[df_train['Age'] < i]['Survived']))
    
plt.figure(figsize = (13, 5))
plt.plot(cummulate_survival_ratio)
plt.title('Survival rate change depending on range of Age')
plt.xlabel('Range of Age')
plt.ylabel('Survival Rate')
plt.show()
```

- Age의 범위에 따른 생존 확률을 구해보니, 젊은 연령대에서의 생존률이 높았고 연령대가 높아질수록 생존률이 급격히 낮아짐을 알 수 있다.
- 이러한 결과가 나온 것으로 보아, **Age 변수는 target 변수를 예측하는데 중요한 feature**라고 생각할 수 있겠다.

### SibSp와 Parch 변수를 하나의 변수(FamilySize)로 합친 후, FamilySize에 따른 생존률의 차이 확인
- SibSp(형제 자매) + Parch(부모, 자녀) = Family(가족)

```python
# 자기 자신도 포함해주어야 하므로 '+ 1' 을 해준다
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1
```

```python
f, ax = plt.subplots(1, 3, figsize = (40, 10))

# FamilySize 변수의 분포 확인
sns.countplot('FamilySize', data = df_train, ax = ax[0])
ax[0].set_title('(1) No. Of Passengers Boarded')

# FamilySize에 따른 생존률의 차이 확인
sns.countplot('FamilySize', hue = 'Survived', data = df_train, ax = ax[1])
ax[1].set_title('(2) Survived countplot depending on FamilySize')
ax[1].legend(loc = 'upper right')

# FamilySize에 따른 생존률 확인
familysize_mean_survived = df_train[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index = True).mean()
familysize_mean_survived.sort_values(by = 'Survived', ascending = False).plot.bar(ax = ax[2])
ax[2].set_title('(3) Survived rate depending on FamilySize')
ax[2].legend(loc = 'upper right')

plt.show()
```

- ***Figure (1)***

    - 대부분 1명(혼자)인 경우가 많다.
- ***Figure (2), Figure (3)***
    - FamilySize에 따른 생존률 차이를 확인해보니, 가족이 4명인 경우가 가장 생존률이 높다. 그리고 가족 수가 3명, 2명인 경우가 그 뒤를 이었다.

- 이러한 결과가 나온 것으로 보아, **FamilySize 변수는 target 변수를 예측하는데 중요한 feature**라고 생각할 수 있겠다.

### Name 변수
- 영어에서는 **Mr, Mrs, Miss**와 같은 **title**이 존재한다.
- 때문에 각 탑승객의 이름에는 위와 같은 **title**이 들어가게 되는데, 이를 사용해서 **"Initial"**이라는 **새로운 변수**를 만들겠다.

```python
# 정규표현식을 사용하여 title을 추출
df_train['Initial'] = df_train.Name.str.extract('([A-Za-z]+)\.')
df_test['Initial'] = df_test.Name.str.extract('([A-Za-z]+)\.')
```

```python
pd.crosstab(df_train['Initial'], df_train['Sex']).T
```

```python
# 위에서 생성된 테이블을 참고하여 남자, 여자가 쓰는 Initial을 구분
df_train['Initial'].replace(['Mlle', 'Mme', 'Ms', 'Dr', 'Major',
                             'Lady', 'Countess', 'Jonkheer', 'Col',
                             'Rev', 'Capt', 'Sir', 'Don', 'Dona'],
                            ['Miss', 'Miss', 'Miss', 'Mr', 'Mr',
                             'Mrs', 'Mrs', 'Other', 'Other',
                             'Other', 'Mr', 'Mr', 'Mr', 'Mr'], inplace = True)

df_test['Initial'].replace(['Mlle', 'Mme', 'Ms', 'Dr', 'Major',
                            'Lady', 'Countess', 'Jonkheer', 'Col',
                            'Rev', 'Capt', 'Sir', 'Don', 'Dona'],
                           ['Miss', 'Miss', 'Miss', 'Mr', 'Mr',
                            'Mrs', 'Mrs', 'Other', 'Other',
                            'Other', 'Mr', 'Mr', 'Mr', 'Mr'], inplace = True)
```

```python
df_train.groupby(['Initial']).mean()
```

**Initial에 따른 생존률**

```python
df_train.groupby('Initial')['Survived'].mean().plot.bar()
```

- 결과를 살펴보면, 여성과 관계있는 **Miss, Mrs가 생존률이 높은 것**을 알 수 있다.
- 이러한 결과가 나온 것으로 보아, **Initial 변수는 target 변수를 예측하는데 중요한 feature**라고 생각할 수 있겠다.
