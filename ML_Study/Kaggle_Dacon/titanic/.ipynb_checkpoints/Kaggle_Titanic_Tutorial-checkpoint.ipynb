# Kaggle - Titanic Tutorial_Part 1
**<분석 목적>**
- 타이타닉에 탑승한 사람들의 신상정보를 활용하여, 승선한 사람들의 **생존 여부**를 **예측**하는 모델을 생성하는 것이다.

**<튜토리얼 과정>**
1. 데이터 셋 확인
    - Null 값 존재 여부를 확인하고, Null 값들에 대한 처리를 해준다.
2. 탐색적 데이터 분석(EDA)
    - 여러 개의 feature들을 개별적으로 분석하고, feature들 간의 상관관계를 확인한다.
    - 또한 시각화 라이브러리들을 활용하여 insight를 도출한다.
3. Feature Engineering(특성 공학)
    - 모델 설정에 앞서, 모델의 성능을 높이기 위한 작업이다.
    - 원-핫-인코딩(One-hot-encoding), class 나누기, 구간 나누기, 텍스트 데이터 처리 등...
4. 모델 생성
    - 사이킷런을 사용해서 모델을 만든다.
5. 모델 학습 및 예측
    - 학습 데이터 셋(train dataset)을 가지고 모델을 학습시킨 후, 테스트 데이터 셋(test dataset)을 가지고 예측을 수행한다.
6. 모델 평가
    - 모델의 예측 성능을 평가한다.
    - 우리가 직면한 문제에 맞게 모델 평가 지표를 결정하면 된다.
        - ex) RMSE, R_square, F1-score, ...

## 탐색적 데이터 분석(EDA)

### Pclass 변수에 따른 생존률의 차이 확인

````python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index = True).count()
````

````python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index = True).sum()
````

- ```sum()```을 사용하면 **각 Pclass별 인원수 중, 생존한(Survived = 1) 사람들의 총합**을 구할 수 있다.

````python
# margins = True 로 설정하면, 표에 'All(총합)' 부분이 추가된다
pd.crosstab(df_train['Pclass'], df_train['Survived'], margins = True).style.background_gradient(cmap = 'summer_r')
````

- ```crosstab()```을 사용하면, 위 결과들을 좀 더 수월하게 볼 수 있다.

````python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index = True).mean()
````

- ```mean()```을 사용하면 **각 Pclass별 생존률**을 구할 수 있다.

````python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index = True).mean().sort_values(by = 'Survived', ascending = False).plot.bar()
````

- Pclass가 좋을수록(1st) 생존률이 높다.

**seaborn의 ```countplot()```을 이용한 시각화**

````python
y_position = 1.02

f, ax = plt.subplots(1, 2, figsize = (18, 8))

df_train['Pclass'].value_counts().plot.bar(color = ['#CD7F32', '#FFDF00', '#D3D3D3'], ax = ax[0])
ax[0].set_title('Number of Passengers By Pclass', y = y_position)
ax[0].set_ylabel('Count')

sns.countplot(x = 'Pclass', hue = 'Survived', data = df_train, ax = ax[1])
ax[1].set_title('Pclass: Survived vs Dead', y = y_position)

plt.show()
````

- 결과를 보면, **Pclass가 높을수록(1st) 생존 확률이 높은 것을 확인**할 수 있다.
- 이 결과를 토대로 나중에 모델 생성 시, **Pclass가 target 변수에 큰 영향을 미친다**고 생각해 볼 수 있다. (즉, **Pclass 변수를 feature로 사용하는 것이 좋을 것**이라는 의미)

### Age 변수에 따른 생존률의 차이 확인

````python
print('최고령 탑승객 : {:.1f} Years'.format(df_train['Age'].max()))
print('최연소 탑승객 : {:.1f} Years'.format(df_train['Age'].min()))
print('탑승객 평균 나이 : {:.1f} Years'.format(df_train['Age'].mean()))
````

````python
fig, ax = plt.subplots(1, 1, figsize = (9, 5))

# kdeplot(): 커널 밀도 함수
sns.kdeplot(df_train[df_train['Survived'] == 1]['Age'], ax = ax)
sns.kdeplot(df_train[df_train['Survived'] == 0]['Age'], ax = ax)

plt.legend(['Survived == 1', 'Survived == 0'])
plt.show()
````

- 보시다시피, 나이가 어린 생존자들이 많은 것을 알 수 있다.

````python
plt.figure(figsize = (8, 6))

df_train['Age'][df_train['Pclass'] == 1].plot(kind = 'kde')
df_train['Age'][df_train['Pclass'] == 2].plot(kind = 'kde')
df_train['Age'][df_train['Pclass'] == 3].plot(kind = 'kde')

plt.xlabel('Age')
plt.title('Age Distribution within classes')
plt.legend(['1st Class', '2nd Class', '3rd Class'])
````

- Pclass가 높을수록(1st) 나이 많은 사람의 비중이 커진다.

**나이대(Range of Age)가 변할 때, 생존률이 어떻게 되는지 확인**

````python
cummulate_survival_ratio = []

for i in range(1, 80):
    cummulate_survival_ratio.append(df_train[df_train['Age'] < i]['Survived'].sum() / len(df_train[df_train['Age'] < i]['Survived']))
    
plt.figure(figsize = (7, 7))
plt.plot(cummulate_survival_ratio)
plt.title('Survival rate change depending on range of Age', y = 1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
plt.show()
````

- 결과를 보면, 나이가 어릴수록 생존률이 높은 것을 확인할 수 있다.
- 따라서 **Age 변수도 모델 생성에 있어서 중요한 feature**임을 알 수 있다.

### SibSp와 Parch 변수를 하나의 변수로 합친 후, 합쳐진 변수에 따른 생존률의 차이 확인
- SibSp(형제 자매) + Parch(부모, 자녀) = Family(가족)

````python
# 자신을 포함해야 하기 때문에 1을 더해준다
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1

print("Maximum size of Family: ", df_train['FamilySize'].max())
print("Maximum size of Family: ", df_train['FamilySize'].min())
````

````python
f, ax = plt.subplots(1, 3, figsize = (40, 10))

# FamilySize 변수의 분포 확인
sns.countplot('FamilySize', data = df_train, ax = ax[0])
ax[0].set_title('(1) No. Of Passengers Boarded', y = 1.02)

# FamilySize 변수에 따른 생존률 차이 확인
sns.countplot('FamilySize', hue = 'Survived', data = df_train, ax = ax[1])
ax[1].set_title('(2) Survived countplot depending on FamilySize', y = 1.02)

# FamilySize 변수에 따른 생존률
df_train[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index = True).mean().sort_values(by = 'Survived', ascending = False).plot.bar(ax = ax[2])
ax[2].set_title('(3) Survived rate depending on FamilySize', y = 1.02)

plt.subplots_adjust(wspace = 0.2, hspace = 0.5)
plt.show()
````

- ***Figure (1)*** 
    - FamilySize가 1부터 11까지 존재함을 알 수 있다.
    - 대부분 1명(혼자)인 경우가 많다.
- ***Figure (2), Figure (3)***
    - FamilySize에 따른 생존률 차이를 확인해보니, 가족이 4명인 경우가 가장 생존률이 높다. 그리고 가족 수가 3명, 2명인 경우가 그 뒤를 이었다.
    - 일반적으로 가족 수가 3 ~ 4명인 경우의 생존 확률이 높은 것을 확인할 수 있다.

### Fare 변수에 따른 생존률의 차이 확인

````python
fig, ax = plt.subplots(1, 1, figsize = (8, 8))

g = sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness : {:.2f}'.format(df_train['Fare'].skew()), ax = ax)
g = g.legend(loc = 'best')
````

- 결과를 보면 알 수 있듯이, Fare 변수에 대한 분포가 굉장히 **치우친 분포**임을 알 수 있다.
    - 이러한 경우, **outlier(이상치)에 굉장히 민감**하게 반응한다.
- 때문에 **로그 변환**이 절대적으로 필요해 보인다.

**Fare 변수에 대해서 "로그 변환"을 수행**
- **DataFrame의 특정 columns에 공통된 작업(함수)를 적용하고 싶으면**, ```map``` 또는 ```apply```를 사용하면 매우 손쉽게 적용할 수 있다. 
    - 여기서는 파이썬의 **lambda** 함수를 이용해서, 간단한 로그를 적용하는 함수를 ```map``` 또는 ```apply```에 인수로 넣어준다.
    - 이 방법은 매우 유용하니까 꼭 숙지해두자!!

````python
# 테스트 데이터 셋에 있는 NaN 값들을 평균값으로 변경
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean()

# apply 대신 map을 사용해도 결과는 동일
df_train['Fare'] = df_train['Fare'].apply(lambda x: np.log(x) if x > 0 else 0)
df_test['Fare'] = df_test['Fare'].apply(lambda x: np.log(x) if x > 0 else 0)
````

````python
fig, ax = plt.subplots(1, 1, figsize = (8, 8))

g = sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness : {:.2f}'.format(df_train['Fare'].skew()), ax = ax)
g = g.legend(loc = 'best')
````

- **로그 변환**을 수행해주니, **치우침의 정도가 어느 정도 개선**이 되었다.
- 이처럼 모델이 조금 더 좋은 성능을 낼 수 있도록 feature들을 처리해주는 작업을 **feature engineering**이라고 한다.
- **Fare 변수** 또한 우리가 생성할 **모델에 feature로 포함**시키겠다.

### Cabin 변수
- 이 변수는 **NaN 값이 약 80%** 나 되기 때문에, target 변수에 영향을 미칠 중요한 정보를 얻기가 어려워 보인다.
