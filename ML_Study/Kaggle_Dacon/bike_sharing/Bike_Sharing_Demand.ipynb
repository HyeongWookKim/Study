- 이번 대회에서 사용하는 **평가 지표**는 "***RMSLE***" 이다.

2. season
    - 계절
    - 1 = spring, 2 = summer, 3 = fall, 4 = winter

    - 공휴일 여부
    - 0 = Not holiday, 1 = holiday

    - 근무일 여부

    - 날씨

    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
6. temp
    - 온도(Celsius)
7. atemp
    - 체감 온도
8. humidiity
    - 습도
9. windspeed
    - 풍속
10. casual
    - 비회원의 자전거 대여 수량
11. registerd
    - 회원의 자전거 대여 수량
12. count
    - 전체 자전거 대여 수량

## 탐색적 데이터 분석(EDA)

### datetime 변수 분할
- **"년-월-일 시:분:초"**로 이루어진 datetime 변수를 **년, 월, 일, 시, 분, 초** 이렇게 6개로 분할해주겠다.

````python
import datetime

df_train['datetime'] = pd.to_datetime(df_train['datetime'], format = "%Y-%m-%d %H:%M:%S", errors = 'raise')
df_test['datetime'] = pd.to_datetime(df_test['datetime'], format = "%Y-%m-%d %H:%M:%S", errors = 'raise')
````

````python
df_train['year'] = df_train['datetime'].dt.year              # 년
df_train['month'] = df_train['datetime'].dt.month            # 월
df_train['day'] = df_train['datetime'].dt.day                # 일
df_train['hour'] = df_train['datetime'].dt.hour              # 시
df_train['minute'] = df_train['datetime'].dt.minute          # 분
df_train['second'] = df_train['datetime'].dt.second          # 초
df_train['weekday'] = df_train['datetime'].dt.weekday        # 요일 = {0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일}

df_test['year'] = df_test['datetime'].dt.year
df_test['month'] = df_test['datetime'].dt.month
df_test['day'] = df_test['datetime'].dt.day
df_test['hour'] = df_test['datetime'].dt.hour
df_test['minute'] = df_test['datetime'].dt.minute
df_test['second'] = df_test['datetime'].dt.second
df_test['weekday'] = df_test['datetime'].dt.weekday
````

**(Target 변수로 사용할) casual 및 registered 변수의 분포 확인**

````python
from scipy import stats

f, ax = plt.subplots(2, 2)
f.set_size_inches(30, 20)

sns.distplot(df_train['casual'], label = 'Skewness: {:.2f}'.format(df_train['casual'].skew()), ax = ax[0, 0])
stats.probplot(df_train['casual'], dist = 'norm', plot = ax[0][1])
sns.distplot(df_train['registered'], label = 'Skewness: {:.2f}'.format(df_train['registered'].skew()), ax = ax[1, 0])
stats.probplot(df_train['registered'], dist = 'norm', plot = ax[1][1])

ax[0, 0].legend(loc = 'best')
ax[1, 0].legend(loc = 'best')
ax[0, 1].set_title('Probability Plot of Casual')
ax[1, 1].set_title('Probability Plot of Registered')

plt.subplots_adjust(wspace = 0.2, hspace = 0.3)
plt.show()
````

- Target 변수로 사용할 casual, registered 변수들의 분포를 확인해보니, **두 변수 모두 굉장히 치우친 분포**를 보인다.
    - 나중에 **로그 변환**을 통해서 정규 분포에 가까운 형태로 만들어주겠다.

## Feature Engineering(특성 공학)

## 모델 학습/예측/평가
- **Target 변수를 2가지 경우**로 나눠서 **각각의 모델의 예측 값들을 더해서 혼합 모델**을 만들 것이다.
    - **Target 변수가 casual**인 경우
    - **Target 변수가 registered**인 경우
- 또한 **target 변수를 count로 설정한 모델**도 만들어서 **혼합 모델과 성능을 비교**해 볼 것이다.
    - **Target 변수가 count**인 경우
- 분석에 **사용할 모델**은 **Random Forest Regressor**이다.
- 대회 기준 평가 지표는 ***RMSLE*** 이지만, 그 외에 ***RMSE*** 와 ***MAE*** 값도 확인해보겠다.

````python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
````

**평가 지표들(RMSLE, RMSE, MAE)을 구해주는 함수 생성**

````python
# 로그 변환 시, NaN 등의 문제로 log()가 아닌, log1p()를 이용해서 "RMSLE" 계산
from sklearn.metrics import make_scorer

def rmsle(predicted_values, actual_values, convertExp = True):

    if convertExp:
        predicted_values = np.exp(predicted_values),
        actual_values = np.exp(actual_values)
        
    # 넘파이로 배열 형태로 바꿔준다
    predicted_values = np.array(predicted_values)
    actual_values = np.array(actual_values)
    
    # 예측값과 실제 값에 1을 더하고 로그를 씌워준다
    log_predict = np.log1p(predicted_values)
    log_actual = np.log1p(actual_values)
    
    # 위에서 계산한 예측값에서 실제값을 빼주고 제곱을 해준다
    difference = log_predict - log_actual
    difference = np.square(difference)
    
    # 평균을 낸다
    mean_difference = difference.mean()
    
    # 다시 루트를 씌운다
    rmsle = np.sqrt(mean_difference)
    
    return rmsle

rmsle_scorer = make_scorer(rmsle)
rmsle_scorer
````

- 먼저 **학습 데이터 셋(train dataset)**과 **target label(Survived)**을 **분리**하겠다.
- 그 다음 ```train_test_split()```을 사용하여, **학습 데이터 셋과 테스트 데이터 셋을 분할**해준다.

````python
X_features = df_train.drop(['casual', 'registered', 'count'], axis = 1, inplace = False)
y_target_casual = df_train['casual']
y_target_registered = df_train['registered']
y_target_count = df_train['count']
````

**1. Target 변수가 count인 모델**

````python
rf_reg_count = RandomForestRegressor(n_estimators = 100, n_jobs = -1, random_state = 0)

k_fold = KFold(n_splits = 5, random_state = 0)

scores_count = cross_val_score(rf_reg_count, X_features, y_target_count, cv = k_fold, scoring = rmsle_scorer)
score_count = np.round(scores_count.mean(), 4)

print('교차 검증별 RMSLE: ', np.round(scores_count, 4))
print('평균 검증 RMSLE: {}'.format(score_count))
````

**2. Taret 변수가 casual인 모델**

````python
rf_reg_casual = RandomForestRegressor(n_estimators = 100, n_jobs = -1, random_state = 0)

k_fold = KFold(n_splits = 5, random_state = 0)

scores_casual = cross_val_score(rf_reg_casual, X_features, y_target_casual, cv = k_fold, scoring = rmsle_scorer)
score_casual = np.round(scores_casual.mean(), 4)

print('교차 검증별 RMSLE: ', np.round(scores_casual, 4))
print('평균 검증 RMSLE: {}'.format(score_casual))
````

**3. Taret 변수가 registered인 모델**

````python
rf_reg_registered = RandomForestRegressor(n_estimators = 100, n_jobs = -1, random_state = 0)

k_fold = KFold(n_splits = 5, random_state = 0)

scores_registered = cross_val_score(rf_reg_registered, X_features, y_target_registered, cv = k_fold, scoring = rmsle_scorer)
score_registered = np.round(scores_registered.mean(), 4)

print('교차 검증별 RMSLE: ', np.round(scores_registered, 4))
print('평균 검증 RMSLE: {}'.format(score_registered))
````
