## 워드 클라우드(WordCloud)
- 워드 클라우드는 **출현 빈도가 높은 단어를 크게 보여줌**으로써 직관적으로 텍스트를 알리는데 도움을 준다
- cmd 명령 프롬프트 창에서 ```pip install wordcloud``` 명령으로 WordCloud를 설치해주어야 한다

````python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
````

**<워드 클라우드 튜토리얼 1> - "이상한 나라의 앨리스" 문서**
- 간단한 워드 클라우드 튜토리얼을 수행해보면서 실습을 진행해보겠다

- 갖고 있는 데이터에서 유의미한 단어 토큰만을 선별하기 위해서, 큰 의미가 없는 단어 토큰을 제거하는 작업이 필요하다
    - 큰 의미가 없다라는 것은 **자주 등장하지만, 분석을 수행함에 있어서 큰 도움이 되지 않는 단어들**을 의미한다
    - ex) I, my, me, over, ...

**<워드 클라우드 튜토리얼 2> - "스타워즈의 A New Hope" 문서**
- 이번엔 다른 텍스트 데이터로 간단한 워드 클라우드 튜토리얼을 수행해보겠다

**NLTK를 통해서 불용어(Stopword) 제거**
- "int", "ext"를 불용어로 보고 제거해주었다

````python
stopwords = set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')
````

- 마찬가지로 **워드 클라우드**를 준비한다

````python
wc = WordCloud(max_words = 1000, mask = mask, stopwords = stopwords,
               margin = 10, random_state = 1).generate(text)
default_colors = wc.to_array()
````

- 앞서 수행해 본 앨리스와는 약간 달리, 색상을 전체적으로 **회색(grey)**으로 채우도록 ```grey_color_func()``` 함수를 생성해주겠다

````python
import random

def grey_color_func(word, font_size, position, orientation,
                    random_state = None, **kwargs):
    return 'hsl(0, 0%%, %d%%)' % random.randint(60, 100)
````

- 이제 워드 클라우드 결과를 시각화 해보겠다!!
    - 아래와 같이 텍스트 분석 주제에 맞는 그림으로 시각화를 하면, 더 좋은 효과를 얻을 수 있다

````python
plt.figure(figsize = (12, 12))
plt.imshow(wc.recolor(color_func = grey_color_func, random_state = 3),
           interpolation = 'bilinear')
plt.axis('off')
plt.show()
````

## Naive Bayes Classifier의 이해 - 영문
- **지도학습**의 한 종류인 "***나이브 베이즈***" 분류기를 **영문으로 된 문서**에 적용해보는 예제를 수행해보겠다

````python
from nltk.tokenize import word_tokenize
import nltk
````

````python
# 연습용 데이터 생성
train = [('i like you', 'pos'),
         ('i hate you', 'neg'),
         ('you like me', 'neg'),
         ('i like her', 'pos')]
````

- 연습용(train)으로 생성한 4개의 문장에서 사용된 전체 단어를 찾기
    - 찾은 단어들을 '말뭉치(corpus)'라고 하겠다

````python
# set()을 사용하여 중복된 값들은 하나만 출력하도록!
all_words = set(word.lower() for sentence in train
                               for word in word_tokenize(sentence[0]))
all_words
````

- 위에서 만들어진 '말뭉치'를 기준으로 train 문장에 속한 단어인지 아닌지를 기록
    - 즉, train의 첫 문장인 'i like you'의 경우 말뭉치 단어 [hate, her, i, like, me, you]를 기준으로 해당 단어가 있는지 없는지를 기록해둔 것이다
    - 아래 **코드의 결과 첫 부분을 해석**해보면, train 문장의 첫 문장에는 **"hate, me, her"**은 **없고(False)** **"i, you, like"**는 **있다(True)**는 뜻이다 

````python
t = [({word: (word in word_tokenize(x[0])) for word in all_words}, x[1])
                                             for x in train]
t
````

- 이제 Naive Bayes 분류기를 동작시켜 보겠다
    - train 문장에 붙은 긍정(pos)/부정(neg) 태그를 이용해서 분류를 수행했다
    - **'hate'라는 단어가 없을 때(False), 긍정(pos)일 비율이 1.7:1**이라는 의미이다
    - **'like'라는 단어가 총 3번 사용**되었고, **그 중 2개의 문장이 긍정(pos)**의 의미였다
        - 때문에 **'like'가 있을 때(True), 긍정(pos)일 비율이 1.7:1**이다

````python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
````

- 이번에는 테스트 문장(test_sentence)을 **'i like MeRui'**로 만들어서 Naive Bayes 분류기를 통과시켜 보겠다
    - **긍정(pos)**이라는 결과가 나왔다

````python
test_sentence = 'i like MeRui'
test_sent_features = {word.lower(): (word in word_tokenize(test_sentence.lower()))
                                     for word in all_words}
test_sent_features
````

````python
classifier.classify(test_sent_features)
````

## Naive Bayes Classifier의 이해 - 한글

- **한글**에서는 **형태소 분석**을 수행해주어야 한다!!

````python
from konlpy.tag import Okt
````

````python
pos_tagger = Okt()
````

- 먼저 **형태소 분석을 수행하지 않은 경우**, 어떤 일이 발생하는지 눈으로 확인해보자!

````python
train = [('메리가 좋아', 'pos'), 
         ('고양이도 좋아', 'pos'),
         ('난 수업이 지루해', 'neg'),
         ('메리는 이쁜 고양이야', 'pos'),
         ('난 마치고 메리랑 놀거야', 'pos')]
````

- ```tokenize()``` 함수를 생성하고, train 문장을 분석

````python
def tokenize(doc):
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm = True, stem = True)]
````

    - 조사(Josa)인지 명사(Noun)인지 구분이 잘 되어 있어서 판독하기 좋다

**이번에는 각 벡터를 정규화 시켜준 후, 거리를 구해보겠다**
- 먼저 각 벡터의 norm을 나눠준 후, 거리를 구하도록 ```dist_norm()``` 함수를 생성

````python
def dist_norm(v1, v2):
    v1_normalized = v1 / sp.linalg.norm(v1.toarray())
    v2_normalized = v2 / sp.linalg.norm(v2.toarray())
    
    delta = v1_normalized - v2_normalized
    
    return sp.linalg.norm(delta.toarray())
````

- 다음으로 **벡터 간 거리** 구하기
    - 거리를 구한 결과가 앞서 수행한 결과와 조금 다르다
    - 그러나 가장 가까운 문장을 찾는 것에 대한 결과는 크게 다르지 않다

````python
best_doc = None
best_dist = 65535
best_i = None

for i in range(0, num_samples):
    post_vec = X.getrow(i)
    d = dist_norm(post_vec, new_post_vec)
    
    print("== Post %i with dist = %.2f : %s" % (i, d, contents[i]))
    
    if d < best_dist:
        best_dist = d
        best_i = i
````

**이번에는 ```tfidf```라는 개념을 적용하여 자연어 처리 수행을 해보도록 하겠다**
- ***tf(term frequency)***와 ***idf(inverse document frequency)***는 텍스트 마이닝에서 사용하는 일종의 **단어별로 부과하는 가중치**이다
- **tf**는 **어떤 단어가 문서 내에서 자주 등장할수록 중요도가 높을 것**으로 보는 것이다
- **idf**는 **비교하는 모든 문서에 만약 같은 단어가 있다면**, 이 단어는 핵심 어휘일지는 모르지만 **문서 간의 비교에서는 중요한 단어가 아니라는 뜻**으로 보는 것이다

````python
import numpy as np

def tfidf(t, d, D):
    tf = float(d.count(t)) / sum(d.count(w) for w in set(d))
    idf = np.lib.scimath.log(float(len(D)) / len([doc for doc in D if t in doc]))
    return tf, idf
````

- 연습용 코드를 보면, 모든 문장에 a가 있기 때문에 idf의 결과는 0이다

````python
a, abb, abc = ['a'], ['a', 'b', 'b'], ['a', 'b', 'c']
D = [a, abb, abc]

print(tfidf('a', a, D))
print(tfidf('b', abb, D))
print(tfidf('a', abc, D))
print(tfidf('b', abc, D))
print(tfidf('c', abc, D))
````

- 이제 위 **두 값을 곱한 것**을 ```tfidf()``` 함수로 수정해서 사용하면 된다

**이번에는 자연어 처리에서 word2vec을 지원하는 gensim을 사용해보겠다**
- 먼저 **조사나 어미 등을 제거**하는 과정을 수행해주었다

````python
import gensim
from gensim.models import word2vec
````

````python
twitter = Okt()
results = []
lines = present_candi_text

for line in lines:
    malist = twitter.pos(line, norm = True, stem = True)
    r = []
    
    for word in malist:
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
            r.append(word[0])
            
    r1 = (" ".join(r).strip())
    results.append(r1)
    print(r1)
````

- 위에서 만들어진 데이터를 혹시 모르니 저장하겠다
- 그리고 ```Word2Vec```을 실행하고, 그 결과를 저장하겠다

````python
data_file = 'pres_girl.data'
with open(data_file, 'w', encoding = 'utf-8') as fp:
    fp.write('\n'.join(results))
````

````python
data = word2vec.LineSentence(data_file)
model = word2vec.Word2Vec(data, size = 200, window = 10, hs = 1, min_count = 2, sg = 1)
model.save('pres_girl.model')
````
