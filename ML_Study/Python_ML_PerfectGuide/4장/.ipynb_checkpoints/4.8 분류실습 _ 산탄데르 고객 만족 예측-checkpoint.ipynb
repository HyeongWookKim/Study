### 데이터 전처리

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib

cust_df = pd.read_csv("./train_santander.csv", encoding = 'latin-1')
print('dataset shape:', cust_df.shape)
cust_df.head(3)
```

```python
cust_df.info()
```

```python
# target(0 또는 1)의 분포 확인
print(cust_df['TARGET'].value_counts())

# target = 1(불만족) 비율 확인
unsatisfied_cnt = cust_df[cust_df['TARGET'] == 1]['TARGET'].count()
total_cnt = cust_df['TARGET'].count()

print('unsatisfied 비율은 {0:.2f}'.format((unsatisfied_cnt / total_cnt)))
```

```python
cust_df.describe( )
```

```python
print(cust_df['var3'].value_counts( )[:10])
```

```python
# var3 피처 값 중, -999999 값은 가장 값이 많은 2 값으로 대체
cust_df['var3'].replace(-999999, 2, inplace = True)
# ID 피처는 단순 식별자에 불과하므로  피처를 드롭
cust_df.drop('ID', axis = 1, inplace = True)

# 피처 세트와 레이블 세트 분리
# 레이블 컬럼은 DataFrame의 맨 마지막에 위치해 컬럼 위치 -1로 분리
X_features = cust_df.iloc[:, :-1]
y_labels = cust_df.iloc[:, -1]
print('피처 데이터 shape:{0}'.format(X_features.shape))
```

```python
from sklearn.model_selection import train_test_split

# train_test_split으로 데이터 셋을 분할할 때, 'stratify = y_labels'를 입력하면 '층화 k-fold'를 수행
# 그럴 경우, '학습 세트 레이블 값 분포 비율'과 '테스트 세트 레이블 값 분포 비율'은 거의 비슷해진다
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels,
                                                    test_size = 0.2, random_state = 0)
train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습 세트 Shape:{0}, 테스트 세트 Shape:{1}'.format(X_train.shape, X_test.shape))
print()

print(' 학습 세트 레이블 값 분포 비율')
print(y_train.value_counts() / train_cnt)
print('\n 테스트 세트 레이블 값 분포 비율')
print(y_test.value_counts() / test_cnt)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# n_estimators는 500으로, random state는 예제 수행 시마다 동일 예측 결과를 위해 설정 
xgb_clf = XGBClassifier(n_estimators = 500, random_state = 156)

# 성능 평가 지표를 'auc'로, 조기 중단 파라미터는 100 으로 설정하고 학습을 수행
# 또한 이 예제에서는 평가를 위한 데이터 셋으로 테스트 데이터 셋을 사용했지만, 이것은 바람직하지 못하다
# 테스트 데이터 셋는 학습 시에는 완전히 알려지지 않은 데이터 셋(검증 데이터 셋)을 사용해야 한다 (과적합 될 수 있기 때문)
# 이 예제에서는 데이터 셋에서 class '0'이 '1'에 비해 월등히 많기 때문에
# 테스트 데이터를 평가용으로 사용했음 (원래는 이렇게 하면 안됌!)
xgb_clf.fit(X_train, y_train, early_stopping_rounds = 100,
            eval_metric = "auc", eval_set = [(X_train, y_train), (X_test, y_test)])  # eval_set은 '넘파이'가 아니라 '데이터 프레임'

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1], average = 'macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 100 으로 감소
xgb_clf = XGBClassifier(n_estimators = 100)

# colsample_bytree 옵션을 설정해서 column 개수를 줄이고자 한다
params = {'max_depth':[5, 7] , 'min_child_weight':[1, 3], 'colsample_bytree':[0.5, 0.75]}

# 하이퍼 파라미터 테스트의 수행속도를 향상 시키기 위해 cv 를 지정하지 않음
gridcv = GridSearchCV(xgb_clf, param_grid = params)  # <참고> 교재에서는 cv = 3 으로 설정했음
gridcv.fit(X_train, y_train, early_stopping_rounds = 30, eval_metric = "auc",
           eval_set = [(X_train, y_train), (X_test, y_test)])

# <참고> 교재에서는 최적 파라미터가 colsample_bytree = 0.75, max_depth = 7, min_child_weight = 1로 나옴
# 왜냐하면 cv = 3 으로 설정해주었기 때문이다
print('GridSearchCV 최적 파라미터:', gridcv.best_params_) 

xgb_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:, 1], average = 'macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
# GridSearchCV로 구한 최적 파라미터들로 해당 파라미터들을 설정한 뒤에 (cv = 3 을 설정하지 않은 경우의 최적 파라미터들을 사용함)
# n_estimators는 1000 으로 증가시키고, learning_rate = 0.02 로 감소, reg_alpha = 0.03 으로 추가함
xgb_clf = XGBClassifier(n_estimators = 1000, random_state = 156, learning_rate = 0.02, max_depth = 7, \
                        min_child_weight = 1, colsample_bytree = 0.5, reg_alpha = 0.03)

# evaluation metric을 'auc'로, early stopping은 200 으로 설정하고 학습 수행
xgb_clf.fit(X_train, y_train, early_stopping_rounds = 200, 
            eval_metric = "auc", eval_set = [(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1], average = 'macro')
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1, 1, figsize = (10, 8))
plot_importance(xgb_clf, ax = ax, max_num_features = 20, height = 0.4)


# 결과를 보면, y축에 피처 명들이 출력된 것을 알 수 있다
# 위에서 eval_set=[(X_train, y_train), (X_test, y_test)] 로 설정을 해주었기 때문이다
# 즉, eval_set을 '넘파이'가 아닌 '데이터 프레임'으로 설정해주었기 때문이다
```

### LightGBM 모델 학습과 하이퍼 파라미터 튜닝

```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier(n_estimators = 500)

# 이 예제에서는 평가를 위한 데이터 셋으로 테스트 데이터 셋을 사용했지만, 이것은 바람직하지 못하다
# 테스트 데이터 셋는 학습 시에는 완전히 알려지지 않은 데이터 셋(검증 데이터 셋)을 사용해야 한다 (과적합 될 수 있기 때문)
# 그러나 이 예제에서는 데이터 셋에서 class '0'이 '1'에 비해 월등히 많기 때문에
# 테스트 데이터를 평가용으로 사용했음 (원래는 이렇게 하면 안됌!)
evals = [(X_test, y_test)]

lgbm_clf.fit(X_train, y_train, early_stopping_rounds = 100, eval_metric = "auc", 
             eval_set = evals, verbose = True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:, 1], average = 'macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 200 으로 감소
LGBM_clf = LGBMClassifier(n_estimators = 200)

params = {'num_leaves':[32, 64],
          'max_depth':[128, 160],
          'min_child_samples':[60, 100],
          'subsample':[0.8, 1]}


# 하이퍼 파라미터 테스트의 수행속도를 향상 시키기 위해 cv 를 지정하지 않음
gridcv = GridSearchCV(lgbm_clf, param_grid = params)  # <참고> 교재에서는 cv = 3 으로 설정했음
gridcv.fit(X_train, y_train, early_stopping_rounds = 30, eval_metric = "auc",
           eval_set = [(X_train, y_train), (X_test, y_test)])

# <참고> 교재에서는 최적 파라미터가 max_depth = 128, min_child_samples = 100, num_leaves = 32, subsample = 0.8 로 나옴
# 왜냐하면 cv = 3 으로 설정해주었기 때문이다
print('GridSearchCV 최적 파라미터:', gridcv.best_params_)

lgbm_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:, 1], average = 'macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

```python
# GridSearchCV로 구한 최적 파라미터들로 해당 파라미터들을 설정한 뒤에 (cv = 3 을 설정하지 않은 경우의 최적 파라미터들을 사용함)
# n_estimators는 1000 으로 증가시킴
lgbm_clf = LGBMClassifier(n_estimators = 1000, num_leaves = 64, sumbsample = 0.8, min_child_samples = 100,
                          max_depth = 128)

# 이 예제에서는 평가를 위한 데이터 셋으로 테스트 데이터 셋을 사용했지만, 이것은 바람직하지 못하다
# 테스트 데이터 셋는 학습 시에는 완전히 알려지지 않은 데이터 셋(검증 데이터 셋)을 사용해야 한다 (과적합 될 수 있기 때문)
# 그러나 이 예제에서는 데이터 셋에서 class '0'이 '1'에 비해 월등히 많기 때문에
# 테스트 데이터를 평가용으로 사용했음 (원래는 이렇게 하면 안됌!)
evals = [(X_test, y_test)]

lgbm_clf.fit(X_train, y_train, early_stopping_rounds = 100, eval_metric = "auc", 
             eval_set = evals, verbose = True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:, 1], average = 'macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```
