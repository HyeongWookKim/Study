## 예시: 문자, 숫자가 아닌 데이터를 찾아서, '' 로 대체해라(삭제해라)

```python
import re
string = "(Dave)"
re.sub('[^A-Za-z0-9]', '', string) # 문자, 숫자가 아닌 데이터를 찾아서, '' 로 대체해라(삭제해라)

# 정규표현식을 사용하지 않고 replace() 함수를 사용한 경우의 코드는 아래와 같다
# string.replace('(', '').replace(')', '')
```

- 위 정규 표현식은 일정한 규칙을 가지고 작성되었으나, 필요한 패턴을 직접 만들 수도 있다.

## 반복 ? , \* , +
* ? 는 앞 문자가 0번 또는 1번 표시되는 패턴 (없어도 되고, 한번 있어도 되는 패턴)
* \* 는 앞 문자가 0번 또는 그 이상 반복되는 패턴
* \+ 는 앞 문자가 1번 또는 그 이상 반복되는 패턴

### ? 사용 예

```python
pattern = re.compile('D?A') # 앞에 문자 D가 없거나, 여러 번 반복되고 마지막이 A 인 문자열
```

```python
pattern.search("A") # 앞에 문자 D가 없는 경우
```

```python
pattern.search("DA") # 앞에 문자 D가 하나 있는 경우
```

```python
pattern.search("DDDDDDA") # 앞에 문자 D가 여러 개 있지만, 그 중에서 매칭되는 것은 바로 앞에 있는 D 1개
```

```python
pattern.search("AA") # 앞에 문자 D가 없고 A가 나온 경우
```

### \* 사용 예

```python
pattern = re.compile('D*A') # 앞에 문자 D가 없거나, 여러 번 반복되고 마지막이 A 인 문자열
```

```python
pattern.search("A")
```

```python
pattern.search("DA")
```

```python
pattern.search("DDDDDDDDDDDDDDDDDDDDDDDDDDDDA")
```

```python
pattern = re.compile('AD*A') # 앞에 문자 D가 없거나, 여러 번 반복되고 마지막이 A 인 문자열
```

```python
pattern.search("ADA")
```

```python
pattern.search("ADDDDDDDDDDDDDDDDDDA")
```

### + 사용 예

```python
pattern = re.compile('D+A')
```

```python
pattern.search("A")
```

```python
pattern.search("DA")
```

```python
pattern.search("DDDDDDDDDDDDDDDDDDDDDDDDDDDDA")
```

## 정규 표현식 라이브러리 함수 사용법

### match 와 search 함수
* match : 문자열 **처음부터** 정규식과 매칭되는 패턴을 찾아서 리턴
* search : 문자열 **전체**를 검색해서 정규식과 매칭되는 패턴을 찾아서 리턴

```python
import re
pattern = re.compile('[a-z]+')
```

```python
matched = pattern.match('Dave')
```

```python
searched = pattern.search("Dave")
```

```python
print(matched) # match() 함수를 사용하면, 문자열의 처음이 'D'(대문자)이기 떄문에 출력 결과가 None
```

```python
print(searched) # search() 함수를 사용하면, 문자열 전체를 검색하기 때문에 1~4번 문자열이 매칭됨
```

### findall 함수: 정규표현식과 매칭되는 모든 문자열을 **리스트** 객체로 리턴함

```python
import re
pattern = re.compile('[a-z]+')
findalled = pattern.findall('Game of Life in Python')
```

```python
print(findalled) # 결과 값을 리스트로 반환
```

```python
pattern2 = re.compile('[A-Za-z]+')
```

```python
findalled2 = pattern2.findall('Game of Life in Python')
```

```python
print (findalled2) # 결과 값을 리스트로 반환
```

### findall 함수를 사용해서 정규표현식에 해당되는 문자열이 있는지 없는지 확인하기

```python
import re
pattern = re.compile('[a-z]+')
findalled = pattern.findall('GAME')
if len(findalled) > 0: # 리스트의 길이가 0보다 큰 경우
    print("정규표현식에 맞는 문자열이 존재함")
else:
    print("정규표현식에 맞는 문자열이 존재하지 않음")
```

## 정규표현식을 통해 성별 구분하기

```python
import openpyxl
 
# 엑셀파일 열기
work_book = openpyxl.load_workbook('train.xlsx')
 
# 현재 Active Sheet 얻기
work_sheet = work_book.active

# 정규표현식
pattern = re.compile(' [A-Za-z]+\.')

# work_sheet.rows는 해당 sheet의 모든 행을 객체로 가지고 있음
for each_row in work_sheet.rows:
    print(each_row[3].value)
    print(pattern.findall(each_row[3].value))
    
work_book.close()
```

## 이름 데이터만 보고 성별 관련 정보 알아내기

```python
import openpyxl
 
# 엑셀파일 열기
work_book = openpyxl.load_workbook('train.xlsx')
 
# 현재 Active Sheet 얻기
work_sheet = work_book.active

# 정규표현식
pattern = re.compile(' [A-Za-z]+\.')

# work_sheet.rows는 해당 sheet의 모든 행을 객체로 가지고 있음
for each_row in work_sheet.rows:
    print(each_row[3].value)
    
    if len(pattern.findall(each_row[3].value)) > 0:
        if pattern.findall(each_row[3].value)[0].strip() == 'Mr.':
            print(pattern.findall(each_row[3].value)[0].strip(), '>>', '남성')
        elif pattern.findall(each_row[3].value)[0].strip() in ['Mrs.', 'Ms.', 'Miss.']:
            print(pattern.findall(each_row[3].value)[0].strip(), '>>', '여성')
        else:
            print(pattern.findall(each_row[3].value)[0].strip(), '>>', '기타')
            
work_book.close()
```

```python
import re
import openpyxl

# 1) train.xlsx 파일 로드하고, 각 sheet 생성하기
work_book = openpyxl.load_workbook('train.xlsx')
work_sheet = work_book.active

work_sheet_mr = work_book.create_sheet('남성')
work_sheet_miss = work_book.create_sheet('미혼여성')
work_sheet_mrs = work_book.create_sheet('기혼여성')
work_sheet_others = work_book.create_sheet('기타')
work_sheet_report = work_book.create_sheet('보고서')

# 2) 분석에 필요한 데이터 변수 초기화하기
mr_survived, mr_unsurvived = 0, 0
miss_survived, miss_unsurvived = 0, 0
mrs_survived, mrs_unsurvived = 0, 0
others_survived, others_unsurvived = 0, 0

regex = re.compile(' [A-Za-z]+\.')

# 3) train.xlsx 파일의 train sheet 데이터를 분석해서, 각 sheet에 데이터 복사하기
for each_row in work_sheet.rows:
    print(each_row[3].value)
    data = regex.findall(each_row[3].value)
    
    # 3-1) train sheet의 첫 행은 항목 이름이 있으므로, 이를 각 sheet에 복사하기
    if each_row[0].row == 1:
        work_sheet_mr.append([each_item.value for each_item in each_row])
        work_sheet_miss.append([each_item.value for each_item in each_row])
        work_sheet_mrs.append([each_item.value for each_item in each_row])
        work_sheet_others.append([each_item.value for each_item in each_row])
    else:
        # 3-2) train 쉬트의 각 행의 이름을 분석해서, 알맞은 쉬트에 복사하기
        if len(data) > 0:
            if data[0] == ' Mr.':
                work_sheet_mr.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    mr_survived += 1
                else:
                    mr_unsurvived += 1

            elif data[0] == ' Miss.':
                work_sheet_miss.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    miss_survived += 1
                else:
                    miss_unsurvived += 1

            elif data[0] == ' Mrs.':
                work_sheet_mrs.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    mrs_survived += 1
                else:
                    mrs_unsurvived += 1

            else:
                work_sheet_others.append([each_item.value for each_item in each_row])
                if each_row[1].value == 1:
                    others_survived += 1
                else:
                    others_unsurvived += 1
                    
# 4) 보고서 sheet 만들기
# 4-1) 보고서 sheet의 컬럼 길이 조정
work_sheet_report.column_dimensions['A'].width = 10
work_sheet_report.column_dimensions['B'].width = 10
work_sheet_report.column_dimensions['C'].width = 10
work_sheet_report.column_dimensions['D'].width = 10

# 4-2) 보고서 sheet의 컬럼명 설정
work_sheet_report.append(['', '', '', ''])
work_sheet_report.append(['분류', '생존자수', '사망자수', '생존률'])

# 4-3) 생존률 계산하기
mr_survived_rate = str(round(100 * mr_survived / (mr_survived + mr_unsurvived), 2)) + '%'
miss_survived_rate = str(round(100 * miss_survived / (miss_survived + miss_unsurvived), 2)) + '%'
mrs_survived_rate = str(round(100 * mrs_survived / (mrs_survived + mrs_unsurvived), 2)) + '%'
others_survived_rate = str(round(100 * others_survived / (others_survived + others_unsurvived), 2)) + '%'

work_sheet_report.append(['남성', mr_survived, mr_unsurvived, mr_survived_rate])
work_sheet_report.append(['미혼여성', miss_survived, miss_unsurvived, miss_survived_rate])
work_sheet_report.append(['기혼여성', mrs_survived, mrs_unsurvived, mrs_survived_rate])
work_sheet_report.append(['기타', others_survived, others_unsurvived, others_survived_rate])

# 5) 각 데이터 저장하기
work_book.save('train_analysis.xlsx')
work_book.close()
```

## 실전 예제 1: 크롤링
* 이커머스(지마켓) 베스트100 상품 타이틀/가격 추출하기
    - 총 12개의 상품군을 모두 추출 (즉, G01~G12)

```python
import requests
from bs4 import BeautifulSoup

base_url = 'http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode='
append_url = ['G01', 'G02', 'G03', 'G04',
              'G05', 'G06', 'G07', 'G08',
              'G09', 'G10', 'G11', 'G12']

for url in append_url:
    res = requests.get(base_url + url)

    if res.status_code != 200:
        print('Page Error')
    else:
        soup = BeautifulSoup(res.content, 'html.parser')

        # G마켓 베스트 상품 페이지에는 'div.best-list'가 2개 존재함
        bestlists_tags = soup.select('div.best-list')
        # 따라서 2개의 'div.best-list' 중, 우리가 원하는 것에 해당하는 2번째 것만 가져옴
        bestlist_tags = bestlists_tags[1].select('ul > li') # 'ul > li'만 select하는 이유는 mismatch를 방지하기 위함!

        for idx, bestlist_tag in enumerate(bestlist_tags):
            title = bestlist_tag.select_one('a.itemname')
            price = bestlist_tag.select_one('div.s-price > strong')
            print(idx + 1, title.text.strip(), price.text.strip())
```

## 실전 예제 2: 크롤링 & 크롤링 + 엑셀 파일로 만들기
* 이커머스(지마켓) 베스트100 상품 타이틀, 가격, 링크, 그리고 판매업체 추출하기
    - 판매업체를 추출하려면 해당 상품을 클릭한 뒤에 "판매업체" 부분만 추출해야 함

```python
import requests, openpyxl
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['랭킹', '상품명', '판매가격', '상세링크', '판매업체'])
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 20
excel_sheet.column_dimensions['D'].width = 80
excel_sheet.column_dimensions['E'].width = 20

base_url = 'http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode='
# 총 12개의 상품군 존재하므로, url도 총 12개가 존재함
append_url = ['G01', 'G02', 'G03', 'G04',
              'G05', 'G06', 'G07', 'G08',
              'G09', 'G10', 'G11', 'G12']

for url in tqdm(append_url):
    res = requests.get(base_url + url)

    if res.status_code != 200:
        print('Page Error')
    else:
        soup = BeautifulSoup(res.content, 'html.parser')

        # G마켓 베스트 상품 페이지에는 'div.best-list'가 2개 존재함
        bestlists_tags = soup.select('div.best-list')
        # 따라서 2개의 'div.best-list' 중, 우리가 원하는 것에 해당하는 2번째 것만 가져옴
        bestlist_tags = bestlists_tags[1].select('ul > li') # 'ul > li'만 select하는 이유는 mismatch를 방지하기 위함!

        for idx, bestlist_tag in enumerate(bestlist_tags):
            title = bestlist_tag.select_one('a.itemname')
            price = bestlist_tag.select_one('div.s-price > strong')
            link = title['href']
            
            # '여행(G11)' 상품군의 경우, 링크를 클릭하면 바로 예약 페이지로 넘어가는 경우가 존재함
            # 따라서 except문을 활용하여 '여행(G11)' 상품군에 대한 처리를 따로 수행
            try:
                # 각 상품별 url을 가져와서, 다시 requests와 bs4를 통해 parsing
                res_info = requests.get(link) 
                soup_info = BeautifulSoup(res_info.content, 'html.parser')

                provider = soup_info.select_one('p > a > strong')

                print(idx + 1, title.text.strip(), price.text.strip(), link, provider.text.strip())

                # 엑셀 시트에 출력 결과 넣기
                excel_sheet.append([idx + 1, title.text.strip(), price.text.strip(), link, provider.text.strip()])
                excel_sheet.cell(row = idx + 2, column = 4).hyperlink = link # '상세링크'를 하이퍼링크 처리
            
            except AttributeError:
                print(idx + 1, title.text.strip(), price.text.strip(), link, None) # '판매업체'를 None으로 지정

                # 엑셀 시트에 출력 결과 넣기
                excel_sheet.append([idx + 1, title.text.strip(), price.text.strip(), link, None]) # '판매업체'를 None으로 지정
                excel_sheet.cell(row = idx + 2, column = 4).hyperlink = link # '상세링크'를 하이퍼링크 처리

# 엑셀 시트 꾸미기(선택)
def cell_style(cell_name):
    '''
    <참고> 색상값 찾기: https://material.io/design/color/#tools-for-picking-colors
    '''
    cell = excel_sheet[cell_name] # 셀 선택하기
    cell.alignment = openpyxl.styles.Alignment(horizontal = 'center') # 중앙 정렬하기
    cell.font = openpyxl.styles.Font(color = '01579B') # 폰트 색깔 바꾸기
    
cells = ['A1', 'B1', 'C1', 'D1', 'E1']
for cell in cells:
    cell_style(cell)
            
excel_file.save('G마켓_상품군별_Best100.xlsx')
excel_file.close()
```
